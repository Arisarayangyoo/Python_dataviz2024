````
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
import os #เชื่่อมpath กับ ไฟล์

path_to_folkales = '/content/drive/MyDrive/data_viz_2024_DATA/folktales.csv'
folk_df = pd.read_csv(path_to_folkales)
folk_df
````

````
#ลบค่า null ใน column ที่ต้องการ
folk_df = folk_df.dropna(subset=['description'])
folk_df
````

````
#ลบตัวอักษรพิเศษ
import re

def clear_text(text):
  # Handle potential non-string values
  if not isinstance(text, str):
    return ''  # Or handle non-string values as needed
  # ใช้ regular expression เพื่อแทนที่อักขระที่ไม่ใช่ตัวอักษรด้วยช่องว่าง
  cleaned_text = re.sub(r'[^ก-๙]', '', text)
  return cleaned_text

# นำฟังก์ชัน clear_text ไปใช้กับคอลัมน์ 'description'
folk_df['cleaned_description'] = folk_df['description'].apply(clear_text)

# แสดงผลลัพธ์
folk_df[ 'cleaned_description']
````

````
# prompt: Find the longest and shortest pieces of information.

# 1.นิทานที่ยาวที่สุดและสั้นที่สุด(ไม่นับสัญลักษณ์พิเศษ)
# หาความยาวของแต่ละเรื่อง
folk_df['length'] = folk_df['cleaned_description'].str.len()

# เรียงลำดับตามความยาว
folk_df_sorted = folk_df.sort_values('length')

# เรื่องที่สั้นที่สุด
shortest_story = folk_df_sorted.iloc[0]
print("นิทานเรื่องที่สั้นที่สุดคือ:", shortest_story['title'])
print("ความยาว:", shortest_story['length'], "ตัวอักษร")

# เรื่องที่ยาวที่สุด
longest_story = folk_df_sorted.iloc[-1]
print("\nนิทานเรื่องที่ยาวที่สุด:", longest_story['title'])
print("ความยาว:", longest_story['length'], "ตัวอักษร")
````

````
#ชนิดของนิทาน (type) ที่มากที่สุดและน้อยที่สุด
# นับจำนวนของแต่ละชนิดของนิทาน (type)
type_counts = folk_df['type'].value_counts()

# หาชนิดของนิทานที่มากที่สุด
most_common_type = type_counts.idxmax()
most_common_count = type_counts.max()

# หาชนิดของนิทานที่น้อยที่สุด
least_common_type = type_counts.idxmin()
least_common_count = type_counts.min()

print("ชนิดของนิทานที่มากที่สุด:", most_common_type, "(", most_common_count, "เรื่อง)")
print("ชนิดของนิทานที่น้อยที่สุด:", least_common_type, "(", least_common_count, "เรื่อง)")
````

````
#3.มี spatial ที่ไม่ใช้ไทยกี่เรื่อง
non_thaispatial_count = folk_df[folk_df['spatial'] != 'ไทย'].shape[0]
print(f"\nมีนิทานที่ไม่ได้มี spatial เป็น ไทย ทั้งหมด {non_thaispatial_count} เรื่อง")
````
