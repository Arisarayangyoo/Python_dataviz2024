````
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
import os #เชื่่อมpath กับ ไฟล์

path_to_folkales = '/content/drive/MyDrive/data_viz_2024_DATA/folktales.csv'
folk_df = pd.read_csv(path_to_folkales)
folk_df
````

````
#ลบค่า null ใน column ที่ต้องการ
folk_df = folk_df.dropna(subset=['description'])
folk_df
````

````
#ลบตัวอักษรพิเศษ
import re

def clear_text(text):
  # Handle potential non-string values
  if not isinstance(text, str):
    return ''  # Or handle non-string values as needed
  # ใช้ regular expression เพื่อแทนที่อักขระที่ไม่ใช่ตัวอักษรด้วยช่องว่าง
  cleaned_text = re.sub(r'[^ก-๙]', '', text)
  return cleaned_text

# นำฟังก์ชัน clear_text ไปใช้กับคอลัมน์ 'description'
folk_df['cleaned_description'] = folk_df['description'].apply(clear_text)

# แสดงผลลัพธ์
folk_df[ 'cleaned_description']
````

````
# prompt: Find the longest and shortest pieces of information.

# 1.นิทานที่ยาวที่สุดและสั้นที่สุด(ไม่นับสัญลักษณ์พิเศษ)
# หาความยาวของแต่ละเรื่อง
folk_df['length'] = folk_df['cleaned_description'].str.len()

# เรียงลำดับตามความยาว
folk_df_sorted = folk_df.sort_values('length')

# เรื่องที่สั้นที่สุด
shortest_story = folk_df_sorted.iloc[0]
print("นิทานเรื่องที่สั้นที่สุดคือ:", shortest_story['title'])
print("ความยาว:", shortest_story['length'], "ตัวอักษร")

# เรื่องที่ยาวที่สุด
longest_story = folk_df_sorted.iloc[-1]
print("\nนิทานเรื่องที่ยาวที่สุด:", longest_story['title'])
print("ความยาว:", longest_story['length'], "ตัวอักษร")
````

````
#ชนิดของนิทาน (type) ที่มากที่สุดและน้อยที่สุด
# นับจำนวนของแต่ละชนิดของนิทาน (type)
type_counts = folk_df['type'].value_counts()

# หาชนิดของนิทานที่มากที่สุด
most_common_type = type_counts.idxmax()
most_common_count = type_counts.max()

# หาชนิดของนิทานที่น้อยที่สุด
least_common_type = type_counts.idxmin()
least_common_count = type_counts.min()

print("ชนิดของนิทานที่มากที่สุด:", most_common_type, "(", most_common_count, "เรื่อง)")
print("ชนิดของนิทานที่น้อยที่สุด:", least_common_type, "(", least_common_count, "เรื่อง)")
````

##Quiz 9
-  วนลูปนับว่ามีนิทานที่มีคำว่า แม่หรือพ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
- มีคำดังกล่าวใน นิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
- คนเเต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่หรือพ่อ มากที่สุด

````
# prompt: วนลูปนับว่ามีนิทานที่มีคำว่า แม่หรือพ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง

count_mom_dad = 0
list_story = []
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    count_mom_dad += 1
    list_story.append(each_row['title'])
print(f'มีนิทานที่มีคำว่า แม่หรือพ่อ ในชื่อเรื่อง ทั้งหมด {count_mom_dad} เรื่อง')
print(f'รายชื่อนิทาน: {list_story}')
````

````
# prompt: มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตหรือนิทานประจำถิ่นมากกว่ากัน

count_life = 0
count_local = 0
#สร้าง dictionary เพื่อเก็บจำนวนนิทานที่มีคำว่า พ่อ หรือ เเม่ เอาไว้
creator_counts = {}
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    if each_row['type'] == 'นิทานชีวิต':
      count_life += 1
    elif each_row['type'] == 'นิทานประจำถิ่น':
      count_local += 1

if count_life > count_local:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตมากกว่า')
elif count_local > count_life:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานประจำถิ่นมากกว่า')
else:
  print('มีคำว่าพ่อ หรือแม่ในนิทาน type นิทานชีวิตและนิทานประจำถิ่นเท่ากัน')
````

````
creator_counts
````

````
# prompt: คนเเต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่หรือพ่อ มากที่สุด

creator_counts = {}
for i, each_row in this_data.iterrows():
  if 'แม่' in each_row['title'] or 'พ่อ' in each_row['title']:
    creator = each_row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

max_count = 0
most_frequent_creator = None
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่หรือพ่อ มากที่สุดคือ: {most_frequent_creator} ({max_count} เรื่อง)")
````

````
# prompt: สร้างตารางของ Director Genre Drama ที่มี Votes มากที่สุดและ Rating มากที่สุด  โดยสร้างตารางเต็ม ๆ

# กรองข้อมูลสำหรับ Genre Drama
drama_df = movie_df[movie_df['Genre'] == 'Drama']

# เรียงลำดับตาม Votes และ Rating (จากมากไปน้อย)
sorted_drama_df = drama_df.sort_values(by=['Votes','Rating'], ascending=[False, False])

# สร้างตารางใหม่โดยเลือกเฉพาะคอลัมน์ที่ต้องการ
result_df = sorted_drama_df[['Name','Duration', 'Director', 'Year', 'Rating', 'Votes']]

# แสดงตารางผลลัพธ์
result_df
````

````
# หา Director ที่กำกับหนัง Genre Romance มากที่สุด
# กรองข้อมูลสำหรับ Genre Romance
drama_df = movie_df[movie_df['Genre'] == 'Romance']

# เรียงลำดับตาม Votes และ Rating (จากมากไปน้อย)
sorted_drama_df = drama_df.sort_values(by=['Votes', 'Rating'], ascending=[False, False])

# สร้างตารางใหม่โดยเลือกเฉพาะคอลัมน์ที่ต้องการ
result_df = sorted_drama_df[['Name','Duration', 'Director', 'Year', 'Rating', 'Votes']]

# แสดงตารางผลลัพธ์
result_df
````

````
# prompt: สร้างตารางของ Director ที่มี Votes มากที่สุดและ Rating มากที่สุด ในปี 2021  โดยสร้างตารางเต็ม ๆ

# กรองข้อมูลสำหรับปี 2021
latest_year = movie_df['Year'].max()
movies_2021 = movie_df[movie_df['Year'] == latest_year]

# เรียงลำดับตาม Votes และ Rating (จากมากไปน้อย)
sorted_movies_2021 = movies_2021.sort_values(by=['Votes', 'Rating'], ascending=[False, False])

# แสดงตารางผลลัพธ์ (ตารางเต็ม)
sorted_movies_2021
````
