```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
import os
path = '/content/drive/MyDrive/data_viz_2024_DATA'
data_file_path = os.path.join(path,'test-TidyData_2024.xlsx')
print(data_file_path)
```

```python
data_Tidy = pd.read_excel(data_file_path)
data_Tidy.head()
```

```python
data_station = data_Tidy[['month', 'station', 'Chlo_a', 'Chlo_b']]
data_station
```

###ความแตกต่างของค่าเฉลี่ย ของข้อมูล chlo_a ทุกๆ station ระหว่างปี 2018 และปี 2019

```python
data_2018 = data_station[data_station['month'].str.contains('2018')]
data_2019 = data_station[data_station['month'].str.contains('2019')]

mean_2018 = data_2018.groupby('station')['Chlo_a'].mean()
mean_2019 = data_2019.groupby('station')['Chlo_a'].mean()

difference = mean_2019 - mean_2018
for station in difference.index:
    print("ค่าเฉลี่ยของข้อมูล Chlo_a ระหว่างปี 2018 และ ปี 2019 ใน station ที่", station , "เท่ากับ", difference[station])
```

###สร้างตารางเฉลี่ย chlo_bทุกๆ station ของเเต่ละเดือน

```python
data_2018_chlo_b = data_2018.groupby(['station', 'month'])['Chlo_b'].mean().unstack()
print(data_2018_chlo_b)
```

```python
data_2019_chlo_b = data_2019.groupby(['station', 'month'])['Chlo_b'].mean().unstack()
print(data_2019_chlo_b)
```

```python
data_merge = pd.merge(data_2018_chlo_b,data_2019_chlo_b, left_on='station',right_on='station')
data_merge
```
